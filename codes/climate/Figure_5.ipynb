```python
# load packages
#import calmap, sys
import sys
sys.path.append('../lib/')
import WDL as wp
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.metrics import r2_score
```

```python
# load data
## load WDL
year_start = 1880
year_end = 2012
## load WDL
pred_WDL = pd.read_csv('predictions/pred_WDL.csv', index_col=0).to_numpy()
## load Frechet
pred_Frechet = pd.read_csv('predictions/pred_Frechet.csv', index_col=0).to_numpy()
## load CLR
pred_CLR = pd.read_csv('predictions/pred_CLR.csv', index_col=0).to_numpy()
## load MDN
pred_MDN = pd.read_csv('predictions/pred_MDN.csv', index_col=0).to_numpy()
## load temperatures
df_temp = pd.read_table('../../data/climate/raw/Complete_TAVG_daily.txt', sep=' ', header=None)
df_temp = df_temp[[2, 3, 4, 6]]
df_temp.columns = ['YEAR', 'MONTH', 'DAY', 'TEMP']
```

```python
# calculate threshold
## url: https://edo.jrc.ec.europa.eu/documents/factsheets/factsheet_heatColdWaveIndex.pdf
## average temp: https://data.giss.nasa.gov/gistemp/graphs_v4/
temps_ = df_temp[(df_temp['YEAR'] >= 1981) & (df_temp['YEAR'] <= 2010)]['TEMP']
thresh_cold = np.quantile(temps_, 0.1)
thresh_warm = np.quantile(temps_, 0.9)
```

```python
# calculate the number of days of heatwave and coldwave
heat_ = np.zeros(133)
cold_ = np.zeros(133)
for idx in range(133):
    year = 1880 + idx
    dt = df_temp[df_temp['YEAR']==year]['TEMP']
    heat_[idx] = sum(dt>thresh_warm)
    cold_[idx] = sum(dt<thresh_cold)
```

```python
# calculate the estimation of WDL
heat_WDL = np.zeros(133)
cold_WDL = np.zeros(133)
heat_Frechet = np.zeros(133)
cold_Frechet = np.zeros(133)
heat_CLR = np.zeros(133)
cold_CLR = np.zeros(133)
heat_MDN = np.zeros(133)
cold_MDN = np.zeros(133)
for idx in range(133):
    cold_WDL[idx] = int(365 * np.mean(pred_WDL[idx] < thresh_cold))
    heat_WDL[idx] = int(365 * np.mean(pred_WDL[idx] > thresh_warm))
    cold_Frechet[idx] = int(365 * np.mean(pred_Frechet[idx] < thresh_cold))
    heat_Frechet[idx] = int(365 * np.mean(pred_Frechet[idx] > thresh_warm))
    cold_CLR[idx] = int(365 * np.mean(pred_CLR[idx] < thresh_cold))
    heat_CLR[idx] = int(365 * np.mean(pred_CLR[idx] > thresh_warm))
    cold_MDN[idx] = int(365 * np.mean(pred_MDN[idx] < thresh_cold))
    heat_MDN[idx] = int(365 * np.mean(pred_MDN[idx] > thresh_warm))
```

```python
fig, ax = plt.subplots(2, 5, figsize=(27, 8), sharex=True)

# these are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='white', alpha=0.5)

ax[0, 0].plot(np.arange(1880, 2013), heat_, color='black')
ax[1, 0].plot(np.arange(1880, 2013), cold_, '--', color='black')
ax[1, 0].axvline(x = 1925, color = 'gray', alpha=0.5)
ax[1, 0].axvline(x = 1975, color = 'gray', alpha=0.5)
#ax[0].set_xlabel('Year', fontsize=20)
ax[0, 0].set_ylabel('Number of Days', fontsize=20)
ax[1, 0].set_ylabel('Number of Days', fontsize=20)
ax[0, 0].set_title('Ground Truth', fontsize=20)
ax[0, 0].tick_params(labelsize=18)
ax[1, 0].tick_params(labelsize=18)
ax[0, 0].text(0.95, 0.92, 'Above 90%', transform=ax[0, 0].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)
ax[1, 0].text(0.95, 0.92, 'Below 10%', transform=ax[1, 0].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)

ax[0, 1].plot(np.arange(1880, 2013), heat_WDL, color='black')
ax[1, 1].plot(np.arange(1880, 2013), cold_WDL, '--', color='black')
ax[1, 1].axvline(x = 1925, color = 'gray', alpha=0.5)
ax[1, 1].axvline(x = 1975, color = 'gray', alpha=0.5)
ax[0, 1].set_title('WDL', fontsize=20)
ax[1, 1].tick_params(labelsize=18)
txt_a = r'$R^2_A = $' + "%.2f" % r2_score(heat_, heat_WDL)
txt_b = r'$R^2_B = $' + "%.2f" % r2_score(cold_[45:95], cold_WDL[45:95])
# place a text box in upper left in axes coords
ax[0, 1].text(0.95, 0.92, txt_a, transform=ax[0, 1].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)
ax[1, 1].text(0.95, 0.92, txt_b, transform=ax[1, 1].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)


ax[0, 2].plot(np.arange(1880, 2013), heat_Frechet, color='black')
ax[1, 2].plot(np.arange(1880, 2013), cold_Frechet, '--', color='black')
ax[1, 2].axvline(x = 1925, color = 'gray', alpha=0.5)
ax[1, 2].axvline(x = 1975, color = 'gray', alpha=0.5)
#ax[2].set_xlabel('Year', fontsize=20)
ax[0, 2].set_title('Frechet', fontsize=20)
ax[1, 2].tick_params(labelsize=18)
txt_a = r'$R^2_A = $' + "%.2f" % r2_score(heat_, heat_Frechet)
txt_b = r'$R^2_B = $' + "%.2f" % r2_score(cold_[45:95], cold_Frechet[45:95])
ax[0, 2].text(0.95, 0.92, txt_a, transform=ax[0, 2].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)
ax[1, 2].text(0.95, 0.92, txt_b, transform=ax[1, 2].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)

ax[0, 3].plot(np.arange(1880, 2013), heat_CLR, color='black')
ax[1, 3].plot(np.arange(1880, 2013), cold_CLR, '--', color='black')
ax[1, 3].axvline(x = 1925, color = 'gray', alpha=0.5)
ax[1, 3].axvline(x = 1975, color = 'gray', alpha=0.5)
#ax[3].set_xlabel('Year', fontsize=20)
ax[0, 3].set_title('CLR', fontsize=20)
ax[1, 3].tick_params(labelsize=18)
txt_a = r'$R^2_A = $' + "%.2f" % r2_score(heat_, heat_CLR)
txt_b = r'$R^2_B = $' + "%.2f" % r2_score(cold_[45:95], cold_CLR[45:95])
ax[0, 3].text(0.95, 0.92, txt_a, transform=ax[0, 3].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)
ax[1, 3].text(0.95, 0.92, txt_b, transform=ax[1, 3].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)

ax[0, 4].plot(np.arange(1880, 2013), heat_MDN, color='black')
ax[1, 4].plot(np.arange(1880, 2013), cold_MDN, '--', color='black')
ax[1, 4].axvline(x = 1925, color = 'gray', alpha=0.5)
ax[1, 4].axvline(x = 1975, color = 'gray', alpha=0.5)
#ax[4].set_xlabel('Year', fontsize=20)
ax[0, 4].set_title('MDN', fontsize=20)
ax[1, 4].tick_params(labelsize=18)
txt_a = r'$R^2_A = $' + "%.2f" % r2_score(heat_, heat_MDN)
txt_b = r'$R^2_B = $' + "%.2f" % r2_score(cold_[45:95], cold_MDN[45:95])
ax[0, 4].text(0.95, 0.92, txt_a, transform=ax[0, 4].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)
ax[1, 4].text(0.95, 0.92, txt_b, transform=ax[1, 4].transAxes, fontsize=16,
        verticalalignment='top', horizontalalignment='right', bbox=props)

ax[0, 1].tick_params(labelsize=18)
ax[0, 2].tick_params(labelsize=18)
ax[0, 3].tick_params(labelsize=18)
ax[0, 4].tick_params(labelsize=18)

ax[0, 0].set_ylim([0, 160])
ax[0, 1].set_ylim([0, 160])
ax[0, 2].set_ylim([0, 160])
ax[0, 3].set_ylim([0, 160])
ax[0, 4].set_ylim([0, 160])

ax[1, 0].set_ylim([0, 390])
ax[1, 1].set_ylim([0, 390])
ax[1, 2].set_ylim([0, 390])
ax[1, 3].set_ylim([0, 390])
ax[1, 4].set_ylim([0, 390])

plt.savefig('../../output/Figure_05.pdf', bbox_inches='tight') 
plt.show()
```
