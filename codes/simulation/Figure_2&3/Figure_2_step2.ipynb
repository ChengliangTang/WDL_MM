```python
# import packages
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# load data
par_array = []
for file_name in os.listdir('output/bootstrap_WDL/'):
    if file_name.startswith('par'):
        par_mat = np.load('output/bootstrap_WDL/' + file_name)
        par_array.append(par_mat)
par_array = np.array(par_array)
par_mean = np.mean(par_array, axis=0)
par_sd = np.std(par_array, axis=0)
```

```python
# visualize the prediction
n_evals = 21
LINEWIDTH = 1
x_eval = np.linspace(-1, 1, n_evals)
y_sort = np.linspace(-1, 1, 100)
fig, ax = plt.subplots(5, 3, figsize=(12, 6), sharex=True)
ax[0, 0].plot(x_eval, par_mean[0, :, 0], c='black', linewidth=LINEWIDTH)
ax[0, 0].plot(x_eval, par_mean[0, :, 0] + 1.96 * par_sd[0, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 0].plot(x_eval, par_mean[0, :, 0] - 1.96 * par_sd[0, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 0].plot(y_sort, np.repeat(np.mean(1 / (1 + np.exp(y_sort))), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[0, 0].set_ylim([0.22, 0.78])
ax[0, 0].set_ylabel(r'$\pi_1$', fontsize=20)


ax[0, 1].plot(x_eval, par_mean[1, :, 0], c='black', linewidth=LINEWIDTH)
ax[0, 1].plot(x_eval, par_mean[1, :, 0] + 1.96 * par_sd[1, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 1].plot(x_eval, par_mean[1, :, 0] - 1.96 * par_sd[1, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 1].plot(y_sort, np.repeat(np.mean(1 / (1 + np.exp(y_sort))), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[0, 1].set_ylim([0.22, 0.78])
ax[0, 1].set_yticklabels([])

ax[0, 2].plot(x_eval, par_mean[2, :, 0], c='black', linewidth=LINEWIDTH)
ax[0, 2].plot(x_eval, par_mean[2, :, 0] + 1.96 * par_sd[2, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 2].plot(x_eval, par_mean[2, :, 0] - 1.96 * par_sd[2, :, 0], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[0, 2].plot(y_sort, 1 / (1 + np.exp(y_sort)), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[0, 2].set_ylim([0.22, 0.78])
ax[0, 2].set_yticklabels([])

ax[1, 0].plot(x_eval, par_mean[0, :, 1], c='black', linewidth=LINEWIDTH)
ax[1, 0].plot(x_eval, par_mean[0, :, 1] + 1.96 * par_sd[0, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 0].plot(x_eval, par_mean[0, :, 1] - 1.96 * par_sd[0, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 0].plot(y_sort, y_sort, c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[1, 0].set_ylim([-1.2, 1.2])
ax[1, 0].set_ylabel(r'$\mu_1$', fontsize=20)

ax[1, 1].plot(x_eval, par_mean[1, :, 1], c='black', linewidth=LINEWIDTH)
ax[1, 1].plot(x_eval, par_mean[1, :, 1] + 1.96 * par_sd[1, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 1].plot(x_eval, par_mean[1, :, 1] - 1.96 * par_sd[1, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 1].plot(y_sort, np.repeat(y_sort.mean(), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[1, 1].set_ylim([-1.2, 1.2])
ax[1, 1].set_yticklabels([])

ax[1, 2].plot(x_eval, par_mean[2, :, 1], c='black', linewidth=LINEWIDTH)
ax[1, 2].plot(x_eval, par_mean[2, :, 1] + 1.96 * par_sd[2, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 2].plot(x_eval, par_mean[2, :, 1] - 1.96 * par_sd[2, :, 1], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[1, 2].plot(y_sort, np.repeat(y_sort.mean(), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[1, 2].set_ylim([-1.2, 1.2])
ax[1, 2].set_yticklabels([])

ax[2, 0].plot(x_eval, par_mean[0, :, 2], c='black', linewidth=LINEWIDTH)
ax[2, 0].plot(x_eval, par_mean[0, :, 2] + 1.96 * par_sd[0, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 0].plot(x_eval, par_mean[0, :, 2] - 1.96 * par_sd[0, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 0].plot(y_sort, np.repeat(np.mean(2 * np.abs(y_sort)**2 + 2), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[2, 0].set_ylim([1.6, 4.4])
ax[2, 0].set_ylabel(r'$\mu_2$', fontsize=20)

ax[2, 1].plot(x_eval, par_mean[1, :, 2], c='black', linewidth=LINEWIDTH)
ax[2, 1].plot(x_eval, par_mean[1, :, 2] + 1.96 * par_sd[1, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 1].plot(x_eval, par_mean[1, :, 2] - 1.96 * par_sd[1, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 1].plot(y_sort, 2 * np.abs(y_sort)**2 + 2, c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[2, 1].set_ylim([1.6, 4.4])
ax[2, 1].set_yticklabels([])

ax[2, 2].plot(x_eval, par_mean[2, :, 2], c='black', linewidth=LINEWIDTH)
ax[2, 2].plot(x_eval, par_mean[2, :, 2] + 1.96 * par_sd[2, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 2].plot(x_eval, par_mean[2, :, 2] - 1.96 * par_sd[2, :, 2], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[2, 2].plot(y_sort, np.repeat(np.mean(2 * np.abs(y_sort)**2 + 2), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[2, 2].set_ylim([1.6, 4.4])
ax[2, 2].set_yticklabels([])

ax[3, 0].plot(x_eval, par_mean[0, :, 3], c='black', linewidth=LINEWIDTH)
ax[3, 0].plot(x_eval, par_mean[0, :, 3] + 1.96 * par_sd[0, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 0].plot(x_eval, par_mean[0, :, 3] - 1.96 * par_sd[0, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 0].plot(y_sort, np.repeat(np.mean(np.abs(y_sort) + 0.5), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[3, 0].set_ylim([0.45, 1.55])
ax[3, 0].set_ylabel(r'$\sigma_1$', fontsize=20)

ax[3, 1].plot(x_eval, par_mean[1, :, 3], c='black', linewidth=LINEWIDTH)
ax[3, 1].plot(x_eval, par_mean[1, :, 3] + 1.96 * par_sd[1, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 1].plot(x_eval, par_mean[1, :, 3] - 1.96 * par_sd[1, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 1].plot(y_sort, np.abs(y_sort) + 0.5, c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[3, 1].set_ylim([0.45, 1.55])
ax[3, 1].set_yticklabels([])

ax[3, 2].plot(x_eval, par_mean[2, :, 3], c='black', linewidth=LINEWIDTH)
ax[3, 2].plot(x_eval, par_mean[2, :, 3] + 1.96 * par_sd[2, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 2].plot(x_eval, par_mean[2, :, 3] - 1.96 * par_sd[2, :, 3], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[3, 2].plot(y_sort, np.repeat(np.mean(np.abs(y_sort) + 0.5), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[3, 2].set_ylim([0.45, 1.55])
ax[3, 2].set_yticklabels([])

ax[4, 0].plot(x_eval, par_mean[0, :, 4], c='black', linewidth=LINEWIDTH)
ax[4, 0].plot(x_eval, par_mean[0, :, 4] + 1.96 * par_sd[0, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 0].plot(x_eval, par_mean[0, :, 4] - 1.96 * par_sd[0, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 0].plot(y_sort, np.abs(y_sort) + 0.5, c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[4, 0].set_ylim([0.45, 1.65])
ax[4, 0].set_xlabel(r'$X_1$', fontsize=20)
ax[4, 0].set_ylabel(r'$\sigma_2$', fontsize=20)

ax[4, 1].plot(x_eval, par_mean[1, :, 4], c='black', linewidth=LINEWIDTH)
ax[4, 1].plot(x_eval, par_mean[1, :, 4] + 1.96 * par_sd[1, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 1].plot(x_eval, par_mean[1, :, 4] - 1.96 * par_sd[1, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 1].plot(y_sort, np.repeat(np.mean(np.abs(y_sort) + 0.5), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[4, 1].set_ylim([0.45, 1.65])
ax[4, 1].set_xlabel(r'$X_2$', fontsize=20)
ax[4, 1].set_yticklabels([])

ax[4, 2].plot(x_eval, par_mean[2, :, 4], c='black', linewidth=LINEWIDTH)
ax[4, 2].plot(x_eval, par_mean[2, :, 4] + 1.96 * par_sd[2, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 2].plot(x_eval, par_mean[2, :, 4] - 1.96 * par_sd[2, :, 4], c='black', linewidth=LINEWIDTH, linestyle= '--')
ax[4, 2].plot(y_sort, np.repeat(np.mean(np.abs(y_sort) + 0.5), 100), c='gray', linewidth=LINEWIDTH, alpha=0.5)
ax[4, 2].set_xlabel(r'$X_3$', fontsize=20)
ax[4, 2].set_ylim([0.45, 1.65])
ax[4, 2].set_yticklabels([])
fig.align_ylabels()
plt.savefig('../../../output/Figure_02.pdf', bbox_inches='tight')  
plt.show()
```
