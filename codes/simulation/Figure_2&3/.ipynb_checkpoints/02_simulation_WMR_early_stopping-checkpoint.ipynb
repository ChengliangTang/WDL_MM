```python
# This script runs WMR algorithm over simulated data
```

```python
# import packages
import os
import numpy as np
import pandas as pd
import wamirpy as wp
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# load data
X = pd.read_csv('../../../data/simulation/setting_3/dat_X.csv').to_numpy()
Y = pd.read_csv('../../../data/simulation/setting_3/dat_Y.csv').to_numpy()
loc_cv = pd.read_csv('../../../data/simulation/setting_3/dat_CV.csv').to_numpy().flatten()
```

```python
# visualize the histogram at each location
fig, ax = plt.subplots(5, 5, figsize=(20, 20))
for i in range(5):
    for j in range(5):
        ax[i, j].hist(Y[20 * i + 2*j], bins=20)
        ax[i, j].set_title('Location (' + str(2*i) + ', ' + str(2*j) + ')')
plt.show()
```

```python
# fit the model using cross-validation
K_mix = 2
n_iter = 1000
lr = 1e-1
v_lr = np.array([1] + [lr] * n_iter)
n_dist = Y.shape[0]
n_levs = 100
n_fold = np.max(loc_cv) + 1
q_vec = np.arange(1, n_levs) / n_levs
## transform Y
Q_mat = np.array([np.quantile(Y[i], q_vec) for i in range(n_dist)])
## initialize predictions
alpha_train = np.zeros((n_dist, K_mix))
mu_train = np.zeros((n_dist, K_mix))
sigma_train = np.zeros((n_dist, K_mix))
alpha_pred = np.zeros((n_dist, K_mix))
mu_pred = np.zeros((n_dist, K_mix))
sigma_pred = np.zeros((n_dist, K_mix))
time_start = datetime.now()
res_list = []
print('Start training:', time_start)
for id_fold in range(n_fold):
    print('This is fold ', id_fold)
    X_train = X[loc_cv != id_fold]
    Y_train = Q_mat[loc_cv != id_fold]
    X_val = X[loc_cv == id_fold]
    Y_val = Q_mat[loc_cv == id_fold]
    res = wp.WDL(X_train, Y_train, X_val, Y_val, q_vec=q_vec, K=K_mix, warm_up=1, 
                 max_iter=n_iter, max_depth=1, patience=5, lr=lr, random_state=2021)
    res_list.append(res)
    for k in range(K_mix):
        alpha_train[loc_cv != id_fold, k] = wp.pred_boost(X_train, res['alpha'][k], lr_=v_lr, n_term=res['iter_best'])
        mu_train[loc_cv != id_fold, k] = wp.pred_boost(X_train, res['mu'][k], lr_=v_lr, n_term=res['iter_best'])
        sigma_train[loc_cv != id_fold, k] = np.exp(wp.pred_boost(X_train, res['sigma'][k], lr_=v_lr, n_term=res['iter_best']))
        alpha_pred[loc_cv == id_fold, k] = wp.pred_boost(X_val, res['alpha'][k], lr_=v_lr, n_term=res['iter_best'])
        mu_pred[loc_cv == id_fold, k] = wp.pred_boost(X_val, res['mu'][k], lr_=v_lr, n_term=res['iter_best'])
        sigma_pred[loc_cv == id_fold, k] = np.exp(wp.pred_boost(X_val, res['sigma'][k], lr_=v_lr, n_term=res['iter_best']))
print('Time:', datetime.now() - time_start)   
```

```python
# print train-val loss on the first fold
plt.plot(res_list[1]['train_loss'], label='train')
plt.plot(res_list[1]['val_loss'], label='test')
#plt.ylim([0, 0.5])
plt.title('Wasserstein Loss vs. Steps')
plt.xlabel('Steps')
plt.ylabel('Wasserstein Loss')
plt.legend()
plt.show()
```

```python
## calculate R-square
stp = n_iter - 1
train_loss = np.mean([res_list[i]['train_loss'][res_list[i]['iter_best']] for i in range(K_mix)])
val_loss = np.mean([res_list[i]['val_loss'][res_list[i]['iter_best']] for i in range(K_mix)])
var_y = np.mean((Q_mat - np.mean(Q_mat, axis=0))**2)
print('Output variance:', var_y)
r_square_train = 1 - train_loss/var_y
r_square_test = 1 - val_loss/var_y
print('Validation loss:', val_loss)
print('Train R-squared:', r_square_train) 
print('Test R-squared:', r_square_test)
```

```python
# Visualize the predicted PDF
pi_train = np.exp(alpha_train)
pi_train = (pi_train.T / np.sum(pi_train, axis=1)).T
pi_pred = np.exp(alpha_pred)
pi_pred = (pi_pred.T / np.sum(pi_pred, axis=1)).T
fig, ax = plt.subplots(2, 5, figsize=(25, 7))
for i in range(2):
    for j in range(5):
        id_loc = 5*(i * 5 + j + 1) - 1
        n_sample = len(Y[id_loc])
        pdf_train = wp.dgmm1d(Y[id_loc], mu_train[id_loc], sigma_train[id_loc], pi_train[id_loc])
        pdf_val = wp.dgmm1d(Y[id_loc], mu_pred[id_loc], sigma_pred[id_loc], pi_pred[id_loc])
        ax[i, j].plot(Y[id_loc], pdf_train, label='Fitted', c='red')
        ax[i, j].plot(Y[id_loc], pdf_val, label='Fitted', c='orange')
        #ax[i, j].set_title(str(id_loc)) 
        ax[i, j].hist(Y[id_loc], bins=20, density=True, color='skyblue', edgecolor='blue')
        ax[i, j].grid()
plt.show()
```

```python
# create partial dependence plot (Quantile)
n_evals = 20
q_evals = np.arange(n_evals+1) / n_evals
p_levs = [0.1, 0.3, 0.5, 0.7, 0.9]
res_mat =  np.zeros((3, (n_evals + 1), 5))
time_start = datetime.now()
print('Start:', time_start)
for i in range(3):
    for j in range(n_evals+1):
        X_c = X.copy()
        X_c[:, i] = np.quantile(X[:, i], q_evals[j])
        alpha_pred = np.zeros((n_dist, K_mix))
        mu_pred = np.zeros((n_dist, K_mix))
        sigma_pred = np.zeros((n_dist, K_mix))
        for k in range(K_mix):
            alpha_pred[:, k] = wp.pred_boost(X_c, res['alpha'][k], lr_=v_lr)
            mu_pred[:, k] = wp.pred_boost(X_c, res['mu'][k], lr_=v_lr)
            sigma_pred[:, k] = np.exp(wp.pred_boost(X_c, res['sigma'][k], lr_=v_lr))
        pi_pred = np.exp(alpha_pred)
        pi_pred = (pi_pred.T / np.sum(pi_pred, axis=1)).T
        res_mat[i, j] = np.mean([wp.qgmm1d(p_levs, mu_pred[l], sigma_pred[l], pi_pred[l]) for l in range(n_dist)], axis=0)
print('Time:', datetime.now() - time_start)
# save as .npy
np.save('../output/res_WMR_qt.npy', res_mat)
```

```python
# create partial dependence plot (CDF)
n_evals = 20
q_evals = np.arange(n_evals+1) / n_evals
p_levs = [0.1, 0.3, 0.5, 0.7, 0.9]
y_levs = np.quantile(Y.flatten(), p_levs)
cdf_mat =  np.zeros((3, (n_evals + 1), 5))
pdf_mat =  np.zeros((3, (n_evals + 1), 5))
time_start = datetime.now()
print('Start:', time_start)
for i in range(3):
    for j in range(n_evals+1):
        X_c = X.copy()
        X_c[:, i] = np.quantile(X[:, i], q_evals[j])
        alpha_pred = np.zeros((n_dist, K_mix))
        mu_pred = np.zeros((n_dist, K_mix))
        sigma_pred = np.zeros((n_dist, K_mix))
        for k in range(K_mix):
            alpha_pred[:, k] = wp.pred_boost(X_c, res['alpha'][k], lr_=v_lr)
            mu_pred[:, k] = wp.pred_boost(X_c, res['mu'][k], lr_=v_lr)
            sigma_pred[:, k] = np.exp(wp.pred_boost(X_c, res['sigma'][k], lr_=v_lr))
        pi_pred = np.exp(alpha_pred)
        pi_pred = (pi_pred.T / np.sum(pi_pred, axis=1)).T
        cdf_mat[i, j] = np.mean([wp.pgmm1d(y_levs, mu_pred[l], sigma_pred[l], pi_pred[l]) for l in range(n_dist)], axis=0)
        pdf_mat[i, j] = np.mean([wp.dgmm1d(y_levs, mu_pred[l], sigma_pred[l], pi_pred[l]) for l in range(n_dist)], axis=0)
print('Time:', datetime.now() - time_start)
# save as .npy
np.save('../output/res_WMR_cdf.npy', cdf_mat)
np.save('../output/res_WMR_pdf.npy', pdf_mat)
```
